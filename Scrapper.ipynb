```python
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_columns', None) 

soup = BeautifulSoup(open("./airbnb_data.html"), "html.parser")
```

```python
barcelona = soup.find_all('table', {'class' : 'table table-hover table-striped barcelona'})[0]
```

```python
barcelona
```

```python
a_tags = barcelona.find_all('a')
```

```python
a_tags
```

```python
links = []

for tag in a_tags:
    links.append(tag['href'])
```

```python
links[0]
```

```python
years = set(['2020', '2019', '2018', '2017'])
links = [link for link in links if not any(year in link for year in years)]
```

```python
links
```

```python
import re
from datetime import datetime
import os
import subprocess

def subprocess_cmd(command):
    print(command)
    process = subprocess.Popen(command,stdout=subprocess.PIPE, shell=True)
    proc_stdout = process.communicate()[0].strip()
    print (proc_stdout)

def download_info(link):
    date = re.search(r'\d{4}-\d{2}-\d{2}', link).group(0)
    if ('listings.csv.gz' in link):
        subprocess_cmd('cd /Users/javi/Desktop/Universidad/4/Visual\ Analytics/Proyecto/data/listings; mkdir -p {0}; cd {0}; wget {1}'.format(date, link))
```

```python
for link in links:
    download_info(link)
```

```python
df = None

for subdir, dirs, files in os.walk('./data/listings/'):
    for file in files:
        if file.endswith(".gz"):
            if df is None:
                df = pd.read_csv(os.path.join(subdir, file), compression='gzip', quotechar='"', error_bad_lines=False)
                df['date'] = df['scrape_id'].astype(str).str[:8]
                df['date'] = pd.to_datetime(df['date'], format='%Y%m%d') 
            else:
                temp = pd.read_csv(os.path.join(subdir, file), compression='gzip', quotechar='"', error_bad_lines=False)
                temp['date'] = temp['scrape_id'].astype(str).str[:8]
                temp['date'] = pd.to_datetime(temp['date'], format='%Y%m%d') 
                df = df.append(temp, sort=False)
```
